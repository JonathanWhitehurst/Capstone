```python


# Reading ESPN's html table of Points/Per Game by school dataset
# ***2021*** 
import pandas as pd
import requests

url = "https://www.espn.com/college-football/stats/team/_/season/2021/table/passing/sort/totalPointsPerGame/dir/desc"

r = requests.get(url)
df_list = pd.read_html(r.text, header=0) # this parses all the tables in webpages to a list

df0=df_list[0]
df1=df_list[1]

#this is because ESPN messed up (data reads as two separate lists with misaligned headers/indexing)
df2 = df1.drop(df1.index[0])
df2 = df2.sort_index().reset_index(drop=True)

df=pd.concat([df0,df2],axis=1)

#Modifying header name
dict = {
        'Points.1': 'Points/G'
}
df.rename(columns=dict,
          inplace=True)


df=df.drop(columns=['Unnamed: 0', 'Total', 'Total.1', 'Passing', 'Passing.1', 'Rushing', 'Rushing.1', 'Points'])
df["Year"]="2021"
print(df)

#df.to_csv("OffPPG2021.csv", index=False)









```

```python
#Replace player positions with Offense/Defense
import pandas as pd

PTP22=pd.read_csv("C:/Users\Jonathan Whitehurst\Code\SavvyCoders\Capstone\playerTransferPortal2021.csv", delimiter="|")

positionType=[]
for x in PTP22['position']:
    if (x=="DL" or x=="CB" or x=="EDGE" or x=="S" or x=="LB" or x=="SDE" or x=="OLB" or x=="ILB" or x=="WDE" or x=="DT"):
        positionType.append("Defense")
    elif (x=="QB" or x=="WR" or x=="RB" or x=="OT" or x=="TE" or x=="IOL" or x=="OL" or x=="OG" or x=="OC" or x=="PRO" or x=="DUAL"):
        positionType.append("Offense")
    elif (x=="K" or x=="LS" or x=="P"):
        positionType.append("Special")
    else:
        positionType.append(x)
    

PTP22["positionType"] = positionType
print(PTP22)

PTP22.to_csv("playerTransferPortal2021.csv", index=False)
```
